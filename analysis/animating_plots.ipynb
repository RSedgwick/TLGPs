# Notebook for Making Gifs of the Predictions of Each of The Models at Each Step

```python
relationtype = 'unrelated' #'non-linear_relation_two_observed_six_new' #'linear_relation' #'unrelated'
seed = 1
data_seed = 1
```

```python
path = pl.Path().home()/'TLGP_plots'
filenames = [path / f'predictions_{relationtype}_two_observed_10_new_{n_points}_points_seed_{seed}_dataseed_{data_seed}.png'
             for n_points in range(1, 40)]
```

```python
for i in range(20):
    for filename in filenames:
        if not os.path.exists(filename):
            filenames.remove(filename)
```

```python
import glob
from PIL import Image

# filepaths
fp_in = filenames
fp_out = pl.Path().home()/f'TLGP_gifs/predictions_{relationtype}_two_observed_10_new_points_seed_{seed}_dataseed_{data_seed}.gif'

# https://pillow.readthedocs.io/en/stable/handbook/image-file-formats.html#gif
imgs = (Image.open(f) for f in fp_in)
img = next(imgs)
imgs = [Image.open(f) for f in filenames]
img1 = imgs[0]
img1.save(fp=fp_out, format='GIF', append_images=imgs[1:],
         save_all=True, optimize=False, duration=1000, loop=0)
```
